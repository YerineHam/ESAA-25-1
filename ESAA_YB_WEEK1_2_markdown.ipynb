# **[개념 정리]**

**판다스**: 파이썬에서 데이터 처리를 위해 존재하는 가장 인기 있는 라이브러리
- 파이썬의 리스트, 컬렉션, 넘파이 등의 내부 데이터 뿐 아니라, CSV 등의 파일을 쉽게 DataFrame으로 변경
- 파일을 DataFrame으로 로딩할 수 있는 편리한 API 제공
    - read_csv(), read_table(), read_fwf()

- DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
    - DataFrame은 리스트와 넘파이 ndarray와 다르게 칼럼명을 가짐
    - 일반적으로 DataFrame으로 변환 시 칼럼명 지정
    - 딕셔너리를 DataFrame으로 변환할 시 딕셔너리의 키를 칼럼명으로, 값을 키에 해당하는 칼럼 데이터로 변환함

- DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
    - 넘파이 ndarray로의 변환: DataFrame 객체의 values를 이용해 쉽게 할 수 있음
    - 리스트로의 변환: values로 얻은 ndarray에 tolist() 호출
    - 딕셔너리로의 변환: DataFrame 객체의 to_dict() 호출

- 칼럼 데이터 세트 생성 및 수정은 '[ ]' 연산자 이용
- 데이터의 삭제는 drop() 이용
    - axis=1을 입력하면 칼럼 축 방향, axis=0을 입력하면 로우 축 방향으로 작업 수행
    - Inplace=False로 두면 자기 자신의 DataFrame의 데이터는 삭제하지 않고, 삭제된 결과를 반환

- Index 객체: DataFrame, Series의 레코드를 고유하게 식별하는 객체
    - reset_index()를 수행하면 새롭게 인덱스를 연속 숫자형으로 할당, 기존 인덱스는 'index'라는 새로운 칼럼명으로 추가
        - 인덱스가 연속된 int 숫자형 데이터가 아닐 경우에 다시 이를 연속 int 숫자형 데이터로 만들 때 주로 사용

- '[ ]' 연산자
    - 슬라이싱 가능(추천 안함), 불린 인덱싱 가능

- iloc[], loc[] 연산자를 통해 데이터를 추출
    - iloc[]: 위치 기반 인덱싱 방식
        - 행 위치와 열 위치를 좌표 정숫값 형태로 입력하여 해당 위치에 있는 데이터를 가져옴
    - loc[]: 명칭 기반 인덱싱 방식
        - 행 위치에 인덱스 값, 열 위치에 칼럼명 입력
        - 슬라이싱 기호를 적용하면 종료 값-1이 아니라 종료 값까지 포함하는 것을 의미함 주의!!
- 불린 인덱싱: 원하는 칼럼명만 별도로 추출 가능

- 정렬, Aggregation 함수, GroupBy 적용
    - 정렬: sort_values() 이용
    - Aggregation 함수: min(), max(), sum(), count()
        - DataFrame에서 바로 호출하면 모든 칼럼에 적용
    - groupby(): 입력 파라미터 by에 칼럼을 입력하면 대상 칼럼으로 groupby
        - 여러 개의 aggregation 함수를 적용하려면 agg() 이용

- 결손 데이터 처리하기
    - isna(): NaN 여부 확인
    - fillna(): NaN 값을 다른 값으로 대체

- apply lambda 식으로 데이터 가공
    - lambda: 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식
    - if else를 지원함, if 절의 경우 if 식보다 반환 값을 먼저 기술해야 함!

**apply lambda 식으로 데이터 가공**

```python
def get_square(a):
    return a**2

print('3의 제곱은:', get_square(3))
```

```python
lambda_square=lambda x:x**2
print('3의 제곱은:', lambda_square(3))
```

```python
a=[1,2,3]
squares=map(lambda x:x**2,a)
list(squares)
```

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60
                                                                                else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat=''
    if age <= 5: cat='Baby'
    elif age <= 12: cat='Child'
    elif age <= 18: cat='Teenager'
    elif age <= 25: cat='Student'
    elif age <= 35: cat='Young Adult'
    elif age <= 60: cat='Adult'
    else : cat='Elderly'

    return cat

#lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
#get_category(X)는 입력값으로 'Age' 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
